```python
#Book 3 has a different format than all the other ones. 
#book_text = book_text.split('PROLOGUE')[1] is required instead of book_text = book_text.split('PROLOGUE')[2] to skip all the 
#foreword and acknowledgements 
#function create_sent_ent_df in Functions.py wont't work on book3 ,hence write the same function with the edited index 
def create_sent_ent_df_book3(book):
    """
    Creates the dataframe with two columns 1)sentences in the book 2) the characters in that sentence
    parameters:
        book_text -> Text file of the book 
    returns :
        Dataframe of sentences and characters in that sentence  
    """
    book_text = open(book,encoding = 'utf-8').read()
    #all text prior to the prologue is unnecessary 
    book_text = book_text.split('PROLOGUE')[1]
    #Load our custom ner for GOT characters  
    nlp = spacy.load('got_ner')
    nlp.max_length = 1768660222
    doc = nlp(book_text)
    sent_ent = []
    for sent in doc.sents:
        ent_list = [ent.text for ent in sent.ents]
        sent_ent.append({'Sentence':sent,'Characters':ent_list})
    sent_ent_df = pd.DataFrame(sent_ent)
    sent_ent_df = sent_ent_df[sent_ent_df['Characters'].map(len)>0]
    sent_ent_df['Characters'] = sent_ent_df['Characters'].map(lambda x:[item.split()[0] for item in x])
    sent_entities = sent_ent_df.reset_index(drop=True)
    return sent_entities
```

#### Connections between the Characters 

I wanted to show the path through which the characters are connected most strongly. But the longest path in an undirected acyclic graph is a np-hard problem. So instead went for the least connected path of characters using Djikstra's shortest path algorithm 

```python
import pickle 
G_1 = pickle.load(open("G1.pickle",'rb'))
Adj_list = G_1.adj
```

```python
#Djikstra's Algorithm 
class DJ:
    def __init__(self):
        self.heap = []
        
    # returns a path from the target to start character 
    def shortestpath(self,end,parent,distance):
        nodes = end 
        newlist = {}
        while nodes is not None:
            newlist[nodes] = distance[nodes]
            nodes = parent[nodes] 
        return newlist
    
    #finds shortest path from start_nodes to all nodes and calls the function shortestpath to return a path from target node  
    #to start node 
    def calc(self,start_node,end_node,adj_list):
        parent ={}
        distance ={}
        visited = {}
        for i in adj_list.keys():
            parent[i] = None 
            distance[i] = np.inf
            visited[i] = False
        distance[start_node] = 0
        self.heap.append(start_node)
        while self.heap:
            currNode = self.heap.pop(0)
            if visited[currNode] == True :
                continue
            for i in adj_list[currNode]:
                source = currNode 
                target = i 
                if i == start_node:
                    continue 
                edge_weight = (adj_list[source][i]['weight']) 
                d = (distance[source] + edge_weight)
                if distance[i] == np.inf:
                    distance[i] = d
                    parent[i] = source 
                elif d < (distance[i]):
                    distance[i] = d
                    parent[i] = source 
                self.heap.append(target)
            visited[currNode] = True
        return self.shortestpath(end_node,parent,distance)
```

```python
def print_network(path):
    #Create the edge_list dataframe from the path results 
    size = len(path)
    characters_list = []
    for i in path.keys():
        characters_list.append(i)
    characters_edge_list = []
    for i,j in enumerate(path):
        if i == size - 1 :
            continue 
        temp_tuple = (characters_list[i],characters_list[i+1],path[j])
        characters_edge_list.append(temp_tuple)
        
    #Create graph G    
    G_try = nx.Graph()
    G_try.add_weighted_edges_from(characters_edge_list)
    pos=nx.spring_layout(G_try)
    nx.draw(G_try,with_labels=True,pos=pos,verticalalignment='baseline',horizontalalignment='right')
    
    #add edge weight labels 
    labels = nx.get_edge_attributes(G_try,'weight')
    nx.draw_networkx_edge_labels(G_try,pos,edge_labels=labels)
    
print_network(path_result)
```
